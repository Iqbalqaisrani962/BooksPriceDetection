```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import pandas as pd

myOptions = Options()
myOptions.headless = True
ser_obj = Service("C:\\Users\\PRO BOOK\\drivers\\chromedriver-win64\\chromedriver.exe")
driver = webdriver.Chrome(service=ser_obj, options=myOptions)

Title = []
Rating = []
Selling_Price = []
Delivery_Charges = []
Seller_Rating = []
Shipping_Timing_Rating = []
Discount_percentage = []
Original_Price = []
Chat_Responce_Rate = []

def details():
    try:
       Title.append(driver.find_element(By.XPATH, "//span[@class='pdp-mod-product-badge-title']").text)
    except NoSuchElementException:
        Title.append("Not Found")
    try:
        Rating.append(driver.find_element(By.XPATH, "//*[@id='module_product_review_star_1']/div/a[1]").text)
    except NoSuchElementException:
        Rating.append("Not Found")
    try:
       Selling_Price.append(driver.find_element(By.XPATH, "//*[@id='module_product_price_1']/div/div/span").text)
    except NoSuchElementException:
        Selling_Price.append("Not Found")
    try:
        Delivery_Charges.append(driver.find_element(By.XPATH, "//*[@id='module_seller_delivery']/div/div/div[3]/div/div[1]/div/div[1]/div[2]").text)
    except NoSuchElementException:
        Delivery_Charges.append("Not Found")
    try:
       Seller_Rating.append(driver.find_element(By.XPATH, "//*[@id='module_seller_info']/div/div[2]/div[1]/div[2]").text)
    except NoSuchElementException:
        Seller_Rating.append("Not Found")
    try:
       Shipping_Timing_Rating.append(driver.find_element(By.XPATH, "//*[@id='module_seller_info']/div/div[2]/div[2]/div[2]").text)
    except NoSuchElementException:
       Shipping_Timing_Rating.append("Not Found")
    try:
       Discount_percentage.append(driver.find_element(By.XPATH, "//*[@id='module_product_price_1']/div/div/div/span[2]").text)
    except NoSuchElementException:
        Discount_percentage.append("Not Found")
    try:
        Original_Price.append(driver.find_element(By.XPATH, "//*[@id='module_product_price_1']/div/div/div/span[1]").text)
    except NoSuchElementException:
        Original_Price.append("Not Found")
    try:
       Chat_Responce_Rate.append(driver.find_element(By.XPATH, "//*[@id='module_seller_info']/div/div[2]/div[3]/div[2]").text)
    except NoSuchElementException:
        Chat_Responce_Rate.append('Not Found')

for j in range(1, 6):
    url = "https://www.daraz.pk/catalog/?spm=a2a0e.searchlist.pagination.2.79cc4353g8EoZM&_keyori=ss&from=input&q=books&page="
    driver.get(url + str(j))
    driver.maximize_window()
    images = driver.find_elements(By.XPATH, "//*[@id='id-a-link']/img")
    for i in range(len(images)):
        images = driver.find_elements(By.XPATH, "//*[@id='id-a-link']/img")
        images[i].click()
        details()
        driver.back()
        
data = {'Title':Title, 'Rating':Rating, 'Selling_Price':Selling_Price, 'Deliver_Charges': Delivery_Charges, 'Seller_Rating':Seller_Rating, 
            'Shipping_Timing_Rating':Shipping_Timing_Rating, 'Discount_Percentage':Discount_percentage, 'Original_Price':Original_Price, 'Chat_Responce_Rate':Chat_Responce_Rate}
df = pd.DataFrame(data)
df.to_csv('data1.csv', index = False)
df = pd.read_csv('data1')
df
driver.close()
```
